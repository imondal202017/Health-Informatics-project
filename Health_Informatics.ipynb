```python
#Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
data["gender"] = le.fit_transform(data["gender"])
data.blood_glucose_type = le.fit_transform(data.blood_glucose_type)
data.urinary_glucose = le.fit_transform(data.urinary_glucose)
data.urinary_protein = le.fit_transform(data.urinary_protein)
data.arrhythmia = le.fit_transform(data.arrhythmia)
```

```python
data
```

```python
data.gender = data.gender.astype('category')
data.blood_glucose_type = data.blood_glucose_type.astype('category')
data.urinary_glucose = data.urinary_glucose.astype('category')
data.urinary_protein = data.urinary_protein.astype('category')
data.arrhythmia = data.arrhythmia.astype('category')
```

```python
data.describe()
```

```python
# Standardize the data
scaler = StandardScaler()
X = scaler.fit_transform(data.values)
```

```python
# Initialize the hierarchical clustering algorithm
clustering_h = AgglomerativeClustering(n_clusters=4, linkage='ward')
```

```python
# Fit the algorithm to the data and get the cluster labels
label_h = clustering_h.fit_predict(X)
```

```python
# Add the cluster labels as a new column to the dataset
data['Cluster_h'] = label_h
```

```python
data_k = data.drop('Cluster_h', axis=1)
```

```python
# Train the clustering model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
label_k = kmeans.fit_predict(data_k)
```

```python
# Get cluster labels
data_k['cluster_k'] = label_k
```
